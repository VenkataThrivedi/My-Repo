# Data Pre-processing and cleaning

## Heatmap of Missing Values

```python
df2 = df.copy()
```

```python
display(df2.isnull().sum())

sns.heatmap(df2.isnull(), yticklabels=False, cbar=False)
```

### Dropping Loan_ID column as it is feels unnecessary for the analysis

```python
df2.drop(['Loan_ID'],axis=1,inplace=True)
```

### Handiling missing Gender values with Mode, and encoding the data into neumerical

```python
gender_mode = df2.Gender.mode()[0]

display(gender_mode)

df2['Gender'].fillna(gender_mode,inplace=True)

display(df2['Gender'].unique())

df2['Gender'].replace(['Male','Female'],[0,1],inplace=True)

display(df2['Gender'].unique())
```

### Handling missing of Married values using Gender correlation
- If the missing Data's Gender value is Male, then married = yes
- else married - No

### Encoding the data into numerical

```python
display(df2['Married'].unique())

for index, row in df2.iterrows():
    if pd.isnull(row['Married']):
        if row['Gender'] == 0:
            df2.at[index, 'Married'] = 'Yes'
        else:
            df2.at[index, 'Married'] = 'No'


display(df2.Married.unique())

df2['Married'].replace(['No','Yes'],[0,1],inplace=True)

display(df2.Married.unique())

```

### Handling Missing Values for Dependents
- if the user is Married then filling the missing value with = mode of Dependent of all Married users
- if the user is Un-Married then filling the missing value with = mode of Dependent of all Un-Married users

### Converting into numerical

```python

sns.countplot(x='Dependents',hue='Married',data=df)

dependents_married_yes = df.loc[df['Married'] == 'Yes', 'Dependents']

dependents_married_yes_mode = dependents_married_yes.mode()[0]

dependents_un_married_yes = df.loc[df['Married'] == 'No', 'Dependents']

dependents_un_married_yes_mode = dependents_un_married_yes.mode()[0]

for index, row in df2.iterrows():
    if pd.isnull(row['Dependents']):
        if row['Married'] == 1:
            df2.at[index, 'Dependents'] = dependents_married_yes_mode
        else:
            df2.at[index, 'Dependents'] = dependents_un_married_yes_mode


display(df2.Dependents.unique())

df2['Dependents'].replace(['0','1','2','3+'],[0,1,2,3],inplace=True)

display(df2.Dependents.unique())

```

### Encoding Education Col. to Numerical

```python
display(df2['Education'].unique())

df2['Education'].replace(['Not Graduate','Graduate'],[0,1],inplace=True)

display(df2['Education'].unique())
```

### Handling missing values for Self Employeed and Enoding to numerical

- If the user is Graduate, then filling the missing value with = mean of Self Emplymnt status of all Graduate users
- If the user is Non-Graduate, then filling the missing value with = mean of Self Emplymnt status of all Non-Graduate users

```python
sns.countplot(x='Self_Employed',hue='Education',data=df)

graduate_employee_status = df.loc[df['Education'] == 'Graduate', 'Self_Employed']

graduate_employee_status_mode = graduate_employee_status.mode()[0]

non_graduate_employee_status = df.loc[df['Education'] == 'Not Graduate', 'Self_Employed']

non_graduate_employee_status_mode = non_graduate_employee_status.mode()[0]

for index, row in df2.iterrows():
    if pd.isnull(row['Self_Employed']):
        if row['Education'] == 1:
            df2.at[index, 'Self_Employed'] = graduate_employee_status_mode
        else:
            df2.at[index, 'Self_Employed'] = non_graduate_employee_status_mode

display(df2['Self_Employed'].unique())

df2.replace(['No','Yes'],[0,1],inplace=True)

display(df2['Self_Employed'].unique())
```

### Handling Missing values of LoanAmount with Median

```python
loan_amount_median = df['LoanAmount'].median()

df2['LoanAmount'].fillna(loan_amount_median,inplace=True)

# display(df2['LoanAmount'].unique())
```

### Handling Missing values of Loan_Amount_Term with Median

```python
loan_amount_term_median = df['Loan_Amount_Term'].median()

df2['Loan_Amount_Term'].fillna(loan_amount_term_median,inplace=True)

display(df2['Loan_Amount_Term'].unique())
```

### Handling Missing values of Credit_History

- Condition used
    - Calculates the mean values of 'ApplicantIncome', 'CoapplicantIncome', and 'LoanAmount' for records where 'Credit_History' is 1.
    - Then, for each row, it imputes missing values in the 'Credit_History' column based on whether the sum of corresponding values in 'ApplicantIncome', 'CoapplicantIncome', and 'LoanAmount' exceeds the combined mean of these three features calculated from records with 'Credit_History' equal to 1.

```python

applicant_income_credit_history_1 = df.loc[df['Credit_History'] == 1, 'ApplicantIncome']

mean_applicant_income_credit_history_1 = round(applicant_income_credit_history_1.mean())

coapplicant_income_credit_history_1 = df.loc[df['Credit_History'] == 1, 'CoapplicantIncome']

mean_coapplicant_income_credit_history_1 = round(coapplicant_income_credit_history_1.mean())

loan_amount_credit_history_1 = df.loc[df['Credit_History'] == 1, 'LoanAmount']

mean_loan_amount_credit_history_1 = round(loan_amount_credit_history_1.mean())


mean_total = mean_applicant_income_credit_history_1 + mean_coapplicant_income_credit_history_1 + mean_loan_amount_credit_history_1

for index, row in df2.iterrows():
    if pd.isnull(row['Credit_History']):
        k = row['ApplicantIncome'] + row['CoapplicantIncome'] + row['LoanAmount']
        if k > mean_total:
            df2.at[index, 'Credit_History'] = 1
        else:
            df2.at[index, 'Credit_History'] = 0

display(df2['Credit_History'].unique())
```

### Encoding Property_Area Column

```python
display(df2['Property_Area'].unique())

df2['Property_Area'].replace(['Urban','Semiurban','Rural'],[1,2,3],inplace=True)

display(df2['Property_Area'].unique())
```

### Encoding Loan_Status Column

```python
display(df2['Loan_Status'].unique())

df2['Loan_Status'].replace(['N','Y'],[0,1],inplace=True)

display(df2['Loan_Status'].unique())
```

# Decision Tree

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Assuming 'X' contains the features and 'y' contains the target variable
X = df2.drop('Loan_Status', axis=1)  # Features
y = df2['Loan_Status']  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


display(X_train)
display(X_test)
display(y_train)
display(y_test)
```

```python

# Initialize DecisionTreeClassifier
clf = DecisionTreeClassifier()

# Fit the classifier to the training data
clf.fit(X_train, y_train)

# Predict on the testing data
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
